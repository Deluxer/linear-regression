## Maniputate Functions

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return x**2

quantity = 100

x = np.linspace(-5, 5, num=quantity)
c = 5
y = f(x) - c

fig, ax = plt.subplots()

ax.plot(x, y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')

plt.show()
```
