# Transcendent Functions

## Trigonometric functions

cos(x), sin(x), tan(x)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return np.sin(x)

quantity = 100

x = np.linspace(-10, 10, num=quantity)

y = f(x)

fig, ax = plt.subplots()

ax.plot(x, y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')

plt.show()
```

## Exponentian Functions

f(x) = a^x

f(x) = e^x

e = euler

```python
def f(x):
    return np.e**x

quantity = 100

x = np.linspace(-1, 1, num=quantity)

y = f(x)

fig, ax = plt.subplots()

ax.plot(x, y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')

plt.show()
```

## Logarithm function

$$
\log_b(x) = n \iff x = b^n
$$

b = base

n = exponent

x = result

$$
\log_2(256) = 8
$$

```python
def f(x):
    return np.log2(x)

quantity = 100

x = np.linspace(0.1, 4, num=quantity)

y = f(x)

fig, ax = plt.subplots()

ax.plot(x, y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')

plt.show()
```

## Heavisede ste Function

```python
def H(x):
    y = np.zeros(len(x))
    for idx, x in enumerate(x):
        if x >= 0:
            y[idx] = 1.0
    return y
        

quantity = 1000

x = np.linspace(0.1, 4, num=quantity)

y = H(x)

fig, ax = plt.subplots()

ax.plot(x, y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')

plt.show()
```
