## Lineal funcion

f(x) = mx + b

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(m, x, b):
    return m*x + b

quantity = 100
m = 10
b = 5

x = np.linspace(-10, 10, num=quantity)

y = f(m, x, b)

fig, ax = plt.subplots()

ax.plot(x, y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
plt.show()
```
