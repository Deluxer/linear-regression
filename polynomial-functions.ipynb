## Polynomial Funcions

```python
import matplotlib.pyplot as plt
import numpy as np
```

f(x) = xÂ²

```python
def f(x):
    return x**2

quantity = 100

x = np.linspace(-10, 10, num=quantity)

y = f(x)

fig, ax = plt.subplots()

ax.plot(x, y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
plt.show()
```
